```python
import numpy as np
import pandas as pd
import datetime as dt

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect

from flask import Flask, jsonify
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
conn = engine.connect()
```

```python
query = """
    Select
        *
    From 
        measurement
"""
```

```python
df_m = pd.read_sql(query, conn)
df_m.head()
```

```python
Base = automap_base()
```

```python
Base.prepare(engine, reflect=True)
```

```python
Measurement = Base.classes.measurement
```

```python
Station = Base.classes.station
```

```python
df_m["date"].max()
```

```python
begin_date = '2016-08-23'
```

```python
session = Session(engine)
```

```python
year_m = session.query(Measurement.station, Measurement.date, Measurement.prcp, Measurement.tobs).filter(Measurement.date > begin_date).all()
year_m
```

```python
df_m = pd.DataFrame(year_m)
df_m
```
